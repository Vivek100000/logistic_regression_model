```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
df = pd.read_csv('train_LZDllcl.CSV')
df
```

```python
for i in df.columns:
  df[i].fillna(df[i].mode()[0], inplace=True)
df.isna().sum().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
x_train=df.loc[0:len(df),['education','no_of_trainings','age','previous_year_rating','length_of_service','KPIs_met >80%','awards_won?','avg_training_score']]
x_train['education'] = labelencoder.fit_transform(x_train['education'])
x_train
```

```python
y_train = df['is_promoted']
y_train
```

```python
x_test = pd.read_csv('test.csv')
employ_id = list(x_test.employee_id)
for i in x_test.columns:
  x_test[i].fillna(x_test[i].mode()[0], inplace=True)
labelencoder = LabelEncoder()
x_test=df.loc[0:len(x_test),['education','no_of_trainings','age','previous_year_rating','length_of_service','KPIs_met >80%','awards_won?','avg_training_score']]
x_test['education'] = labelencoder.fit_transform(x_test['education'])
x_test.isna().sum().sum()
x_test
```

```python
clf = RandomForestClassifier(max_depth=8, random_state=42, n_estimators = 200, max_features='log2',criterion='entropy')
x_train
y_train 
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
y_pred
```

```python
y_pred =list(y_pred)
employ_id 
len(y_pred)
```

```python
data = {'employee_id':list(employ_id)}
data = pd.DataFrame(data)
data['is_promoted'] = y_pred[1:len(y_pred):]
data
```
